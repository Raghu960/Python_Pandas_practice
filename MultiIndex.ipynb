```python
fd=pd.read_csv(r"F:\4-1 Files\Udemy - Data Analysis with Pandas and Python\1. Installation and Setup\1.1 pandas.zip\foods.csv")
fd["Gender"]=fd["Gender"].astype("category")
fd["Item"]=fd["Item"].astype("category")
fd.head(3)
```

```python
fd.pivot_table(index="Gender",columns="Item",values="Spend",aggfunc="mean")
```

```python
s=fd.pivot_table(values="Spend",index=["Gender","Item"],columns=["Frequency","City"],aggfunc="sum")
s.head(3)
```

```python
s.unstack()
```
